```python
import os
import matplotlib.pyplot as plt
import numpy as np

folder_path = "filtered_images"

def extract_ctr_values(folder_path):
    ctr_values = []
    for filename in os.listdir(folder_path):
        try:
            ctr_str, _, _ = filename.split("_", 2)
            ctr = float(ctr_str) / 100.0 
            ctr_values.append(ctr)
        except Exception as e:
            print(f"Praleidžiamas failas: {filename} – klaida: {e}")
    return np.array(ctr_values)

ctr_values = extract_ctr_values(folder_path)

print(f"CTR min: {ctr_values.min():.4f}")
print(f"CTR max: {ctr_values.max():.4f}")
print(f"CTR median: {np.median(ctr_values):.4f}")
print(f"CTR 95th percentile: {np.percentile(ctr_values, 95):.4f}")

# Histogramos brėžinys
plt.figure(figsize=(10, 6))
plt.hist(ctr_values, bins=50, color='skyblue', edgecolor='black')
plt.axvline(0.27, color='red', linestyle='--', label='0.27 cutoff')
plt.title("CTR pasiskirstymo histograma")
plt.xlabel("CTR reikšmė")
plt.ylabel("Vaizdų skaičius")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
